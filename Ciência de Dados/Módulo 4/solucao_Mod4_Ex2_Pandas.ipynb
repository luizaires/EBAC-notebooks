## 1) Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)

df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
#Criando um data frame com a distribuição de questões no vestibular por assunto
assuntos = ['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química']
n_questoes = [24, 18, 8, 8, 8, 12, 12]
df_vestibular = pd.DataFrame(data = n_questoes, index = assuntos, columns = ['número de questões'])
df_vestibular
````

````python
#Colocando o número de acertos por assunto para os 100 alunos em único data frame
df_lista = [df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui]
df_notas= pd.concat(df_lista, axis = 1, ignore_index = True)
df_notas.columns = assuntos #Renomeando as colunas, usando a lista assunto
df_notas
````

---

````python
#Parte 1 - Definindo um data frame contendo o percentual de acertos dos alunos por assunto
df_percentual_acertos = df_notas/n_questoes
df_percentual_acertos
````

---

````python
#Parte 2-Calculando total de acertos de cada aluno
#O método sum(), quando aplicado a um data frame, retorna uma série contendo o somatório dos valores da coluna
#indexada pelo nome da coluna.
#Para calcular o total de acertos para cada aluno, devemos somar os valores de cada linha. Portanto, devemos
#utilizar o método T para retornar a transposta, antes de aplicar o método sum()
total_acertos = df_notas.T.sum()
df_notas['Total acertos'] = total_acertos #Criando nova coluna "Total acertos"
df_notas
````

---

````python
#Parte 3-Cálculo do percentual de acertos de cada aluno
total_acertos/sum(n_questoes)
````

---

````python
#Parte 4-Total de alunos que passaram para 2ª fase por obterem mais de 45 acertos
sum(total_acertos > 45)
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
#Parte 1-Data frame contendo os dados apenas dos alunos desqualificados
i_desqualificados = (df_notas['Matemática'] == 0)|(df_notas['Física'] == 0) |(df_notas['Química'] == 0)
df_desqualificados = df_notas[i_desqualificados]
df_desqualificados
````

---

````python
#Parte 2-Contando número de alunos desqualificados, usando método count()
#Se aplicado ao data frame, o método count() retorna uma series, contendo o quantidade de valores em cada coluna.
#Por isso, aplicamos um método a uma linha qualquer para obter um único valor
df_desqualificados['Total acertos'].count()  
````

---

````python
#Parte 3-Média dos alunos em História
df_desqualificados['História'].mean()
````

````python
#Parte 3-Média dos alunos em Geografia
df_desqualificados['Geografia'].mean()
````

---

````python
#Parte 4-Data Frame dos alunos que passaram para segunda fase
df_segunda_fase = df_notas[(~i_desqualificados) & (df_notas['Total acertos'] > 45)]
df_segunda_fase
````
